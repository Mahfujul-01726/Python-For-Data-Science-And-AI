# 1. For Loop

**a) Loop through a list:**

```python
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)
```

```python
names = ['A', 'B']
ages = [20, 30]
for name, age in zip(names, ages):
    print(name, age)
```

**b) Loop through a string:**

```python
for char in "Python":
    print(char)
```

**c) Loop using range() function:**

```python
# Numbers 0 to 4
for i in range(10):
    print(i)

```

```python
# Numbers 2 to 5
for i in range(2, 10):
    print(i)

```

```python
# Even numbers 0 to 8
for i in range(0, 10, 2):
    print(i)
```

```python
import pandas as pd

df = pd.DataFrame({
    'name': ['Alice', 'Bob'],
    'score': [85, 92]
})

# iterrows() gives index, row as Series
for index, row in df.iterrows():
    print(row['name'], row['score'])
```

```python
import numpy as np

arr = np.array([10, 20, 30])
for value in arr:
    print(value)
```

```python
dataset = ["AI", None, "Data Science", "", "ML"]

cleaned = []
for item in dataset:
    if item:
        cleaned.append(item)

print(cleaned)
```

# 2. while Loops

```python
count = 0
while count < 5:
    print(count)
    count += 1
```

```python
i = 0
while True:
    print(i)
    i += 1
    if i >= 5:
        break
```

# 3. Loop Control Statements

**a) break**

```python
for i in range(10):
    if i == 5:
        break
    print(i)
```

**b) continue**

```python
for i in range(5):
    if i == 2:
        continue
    print(i)
```

# 4. Nested Loops

```python
for i in range(1, 4):
    for j in range(1, 4):
        print(f"{i} * {j} = {i*j}")
    print("---")  # Separator
```
